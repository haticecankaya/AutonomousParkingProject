```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv('otonom_park_dataset.csv')

print(df.info())
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(20, 15))
features_to_plot = [
    'r',
    'v',
    'd_min',
    'd_brake',
    'slip_risk',
    'e_lane',
    'batt',
    'score'
]


for i, col in enumerate(features_to_plot):
    plt.subplot(3, 3, i + 1)
    sns.histplot(data=df, x=col, kde=True, bins=30, color='skyblue')
    plt.title(f'{col} Dağılımı (Distribution)')
    plt.xlabel(col)
    plt.ylabel('Frekans')


plt.subplot(3, 3, 9)
sns.countplot(x='label', data=df, palette='viridis')
plt.title('Etiket Dağılımı (0: Güvenli, 1: Riskli)')
plt.xlabel('Label')
plt.ylabel('Sayı (Count)')

plt.tight_layout()
plt.show()
```
